```python
#box plot
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,6))
sns.boxplot(x='HouseNetWorth',y='HousePrice',data=data,palette='pastel')
plt.title('House Price across House Net worth Categories')
plt.show()

```

The HouseNetworth in the high category appears to have outliers.

```python
# Line chart for HousePrice over StoreArea
plt.figure(figsize=(10, 6))
sns.lineplot(x='StoreArea', y='HousePrice', data=data, color='blue')
plt.title('House Price vs Store Area')
plt.xlabel('Store Area')
plt.ylabel('House Price')

# Line chart for HousePrice over BasementArea
sns.lineplot(x='BasementArea', y='HousePrice', data=data, color='red')

# Line chart for HousePrice over LawnArea
sns.lineplot(x='LawnArea', y='HousePrice', data=data, color='green')

# Line chart for HousePrice over HouseNetWorth
sns.lineplot(x='HouseNetWorth', y='HousePrice', data=data, color='orange')

plt.legend(['StoreArea', 'BasementArea', 'LawnArea', 'HouseNetWorth'])
plt.show()
```

As the lawn area increases the  house price also increases.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set style
sns.set_style("whitegrid")

# pairplot with hue
plt.figure(figsize=(10, 8))
sns.pairplot(data, hue="HouseNetWorth", palette="husl")
plt.title('Pairplot with Different Colors based on House Net Worth')
plt.show()
```

The figure explains the relationship between the variables. e.g. as the store area(y) and house price (x) increases,the houseNetworth is high. This implies that the smaller the area,the lower houseprice and a low networth which results to increased affordability.

```python
#histogram to understand the distribution
data.hist(figsize=(12,10),bins=30)
plt.show()
```

the histograms appear to be right-skewed which can be thought of as starting with a symmetric data set and pulling on its right tail. We can appear to pull the graph to the right by adding more values near the maximum of the distribution, which pulls on the mean more strongly than it does either the median or mode.
As a result, the median will be lower than the mean, because the mean is more sensitive to the higher values and is drawn towards the tail of the histogram.

```python


# Create a scatter plot with a regression line
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot

# Create a scatter plot with a regression line
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.regplot(data=data, x='StoreArea', y='HousePrice', marker='o', line_kws={"color": "red"})
sns.regplot(data=data, x='BasementArea', y='HousePrice', marker='o', line_kws={"color": "blue"})
sns.regplot(data=data, x='LawnArea', y='HousePrice', marker='o', line_kws={"color": "green"})

# Add titles and labels
plt.title('Scatter Plot with Linear Fit', fontsize=16)
plt.xlabel(' Area (sq ft)', fontsize=14)
plt.ylabel('House Price ($)', fontsize=14)

# Show the plot
plt.show()

# Show the plot
plt.show()
```

Most the data is clustered tightly around its regression line indicating a strong linear relationship.

```python
#correlation with heatmap to interpret the relationship and multicollinearity
s=data.select_dtypes(include="number").corr()
```

```python
sns.heatmap(s,annot=True)
```

In a correlation heatmap, each variable is represented by a row and a column, and the cells show the correlation between them. The color of each cell represents the strength and direction of the correlation, with darker colors indicating stronger correlations while lighter colors indicate weaker correlations.
Positive correlations (when one variable increases, the other variable tends to increase) are usually represented by warm colors, such as red or orange.
Negative correlations (when one variable increases, the other variable tends to decrease) are usually represented by cool colors, such as blue or green.
 Understanding correlation heatmaps helps us identify patterns and relationships between multiple variables.
 e.g.The correlation coefficient between houseprice and store area is 0.82 indicating a strong positive relationship between the two i.e as the store area increases the house price also increases.
